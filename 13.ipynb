```python
# factorial
num = int(input("enter number "))
factorial = 1
if num < 0 :
  print("factorail for negative integer doesnt exist")
elif num == 0:
  print("factorial is 1")
else:
  for i in range(1,num + 1):
     factorial = factorial * i
  print(f"factorial of {num} is {factorial}")
```

```python
def factorial(num):
  result = 1
  for i in range(1,num+1):
    result *= i
  return result
n = int(input("enter a number "))
print(f"factorial of {n} is {factorial(n)}")
```

```python
  # prime no
def prime(num):
    if num <= 1:
      return False
    for i in range(2,int(num**0.5)+1):
      if num % i == 0:
        return False
    return True
n = int(input("enter no "))
prime(n)
if prime(n):
   print(f"{n} is a prime no")
else:
   print(f"{n} not prime")
```

```python
#palindrome
def palindrome(string):
  return string == string[::-1]
userinput=input("enter string : ")
if palindrome(userinput):
  print("it is palindrome")
else:
  print("it is not palindrome")
```

```python
#min max
def minmax(arr):
  max_val = arr[0]
  min_val = arr[0]
  for num in arr:
    if num > max_val:
      max_val = num
    if num < min_val:
      min_val = num
  return max_val,min_val
array = list(map(int,input("enter no separated by space : ").split()))
max_val , min_val = minmax(array)
print(f"max = {max_val} ,min = {min_val}")
```

```python
def second_largest(arr):
    if len(arr) < 2:
        return "Array should have at least two elements."

    # Initialize first and second to the smallest possible value in the array
    first = second = arr[0]

    for num in arr:
        if num > first:
            second = first
            first = num
        elif num > second and num != first:
            second = num
    return second

# Input
array = list(map(int, input("Enter numbers separated by space: ").split()))
print("Second Largest Element:", second_largest(array))
```

```python
# remove duplicate elements from array
def removeduplicate(arr):
  unique = []
  for num in arr:
    if num not in unique:
      unique.append(num)
  return unique
array = list(map(int,input("enter numbers ").split()))
uniquearr = removeduplicate(array)
print("array without duplicates : ",uniquearr)
```

```python
#sum of all elements in arr
def sum(arr):
  total = 0
  for num in arr:
    total += num
  return total
array = list(map(int,input("enter nos ").split()))
print(sum(array))
```
