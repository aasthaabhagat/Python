```python
def evenodd(numbers):
  even = []
  odd = []
  for num in numbers :
    if num % 2 ==0:
      even.append(num)
    else:
      odd.append(num)
    return even,odd
numbers = [1,2,3,4,5]
evennos , oddnos = evenodd(numbers)
print("even ",evennos)
```
